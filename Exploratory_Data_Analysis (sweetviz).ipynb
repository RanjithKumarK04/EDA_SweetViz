## Powerful EDA (Exploratory Data Analysis)

### Dataset Location:

https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data

```python
# Import Required Libraries:

import pandas as pd
import sweetviz
```

```python
# Import Dataset:

train = pd.read_csv("E:\\house-prices\\train.csv")
test = pd.read_csv("E:\\house-prices\\test.csv")
```

```python
# Training dataset view:

train.tail()
```

### Creating Report using sweetviz.analyze function:

```python
my_report = sweetviz.analyze([train, "Train"], target_feat="SalePrice")
```

```python
# View report in html format:

my_report.show_html(filepath="my_report.html")
```

## Comparing two dataframes: (Train vs Test Data)

```python
my_report_compare = sweetviz.compare([train, 'Train'],[test,'Test'], 'SalePrice')
```

```python
# View report in html format:

my_report_compare.show_html(filepath="my_report_compare.html")
```
